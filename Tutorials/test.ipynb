```python
from gtts import gTTS
from IPython.display import Audio
```

```python
def say(text):
  google_audio = gTTS(text)
  filename = 'google_audio.wav'
  google_audio.save(filename)
  display(Audio(filename, autoplay=True))
  
```

```python
say('Is this working?')
```
