# But what is it?
Artificial Intelligence is creating a program that is not just capable of learning about its enviroment (in whatever that may be) But a program that uses what it learns to make a decision often called a task. While the learning part of A.I is summed up often as Machine Learning. A.I is often a summation of parts, learning and making decisions.
